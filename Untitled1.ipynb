```python
#######train = train_1['game_season'].map(z) ################3
def dist(x):
    if (pd.isnull(x[9])) and (~pd.isnull(x[12])):
        if x[12] == 'Goal Area':
            return 20.0
        elif x[12] == 'Goal Line':
            return 28.0
        elif x[12] == 'Left Corner':
            return 43.0
        elif x[12] == 'Mid Ground Line':
            return 75.0
        elif x[12] == 'Mid Range':
            return 37.0
        elif x[12] == 'Penaltx[12] Spot':
            return 45.0
        elif x[12] == 'Right Corner':
            return 43.0
        
    elif (~pd.isnull(x[9])):
        return x[9]
    else:
        return 0.0
  
```

```python
new = train.apply(dist, axis = 1)
```

```python
train['distance_of_shot'] = new
```
