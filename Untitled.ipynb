```python
def dist(x):
    if (pd.isnull(x[9])) and (~pd.isnull(x[12])):
        if x[12] == 'Goal Area':
            return float(20.0)
        elif x[12] == 'Goal Line':
            return float(28.0)
        elif x[12] == 'Left Corner':
            return float(43.0)
        elif x[12] == 'Mid Ground Line':
            return float(75.0)
        elif x[12] == 'Mid Range':
            return float(37.0)
        elif x[12] == 'Penaltx[12] Spot':
            return float(45.0)
        elif x[12] == 'Right Corner':
            return float(43.0)
        
    elif (~pd.isnull(x[9])):
        return float(x[9])
    else:
        return float(0.0)
  
        
```
